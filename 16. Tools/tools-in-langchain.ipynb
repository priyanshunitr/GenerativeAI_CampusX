### Custom Tool

Method 1 - Using @tool

```python
from langchain_core.tools import tool
```

```python
# step 1: Create a custom tool (doc string helps llm find what the tool does)
def multiply_nums(a, b):
    """Multiply two numbers."""
    return a * b
```

```python
# step 2: add type hints
def multiply_nums(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b
```

```python
# step 3: add tool decoration

@tool
def multiply_nums(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b
```

```python
result = multiply_nums.invoke({"a": 6, "b": 7})
print(result)
```

```python
print(multiply_nums.name)
print(multiply_nums.description)
print(multiply_nums.args)
```

```python
multiply_nums.args_schema.model_json_schema()
```

Method 2 - Using StructuredTool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInputs(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
```

```python
def multiply_func(a: int, b:int) -> int:
    return a * b
```

```python
multply_tool = StructuredTool.from_function(
    name="multiply",
    description="Multiply two numbers",
    func=multiply_func,
    args_schema=MultiplyInputs
)
```

```python
result = multply_tool.invoke({"a": 6, "b": 4})
result
```

```python
print(multply_tool.name)
print(multply_tool.description)
print(multply_tool.args)
```

Using BaseTool class

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# arg_schema for pydantic
class MultiplyInputs(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
```

```python
class MultiplyTool(BaseTool):
    name: str = "Multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInputs
    
    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multply_tool2 = MultiplyTool()
```

```python
result = multply_tool2.invoke({"a": 5, "b": 3})
result
```

```python
print(multply_tool2.name)
print(multply_tool2.description)
print(multply_tool2.args)
```

Toolkit

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
    """A toolkit for basic math operations."""

    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()
tools
```

```python
for tool in tools:
    print(tool.name, "=>", tool.description)
```
