```python
from dotenv import load_dotenv
load_dotenv()  # take environment variables from .env file
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

Tool Creation

```python
# tool creation

@tool
def multiply(a: int, b: int) -> int:
    """Given two numbers a and b, this tool returns their product."""
    return a * b
```

```python
multiply.invoke({"a": 6, "b": 2})
```

```python
multiply.name
```

```python
multiply.description
```

```python
multiply.args
```

Tool Binding - Linking the tool with LLM

```python
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
```

```python
llm_with_tools =llm.bind_tools([multiply])
```

```python
llm_with_tools
```

Tool Calling - Here LLM just tell you which tool to call, it doesnot actaully cal the tool and generate the result

```python
llm_with_tools.invoke("Hii, how are you?")
```

```python
query = HumanMessage(content="What is the product of 24 and 25?")
```

```python
messages = [query]
```

```python
result = llm_with_tools.invoke(messages)
```

```python
result
```

```python
messages.append(result)
messages
```

```python
result.tool_calls[0]
```

Tool Execution

```python
tool_result = multiply.invoke(result.tool_calls[0])
```

```python
tool_result
```

```python
messages.append(tool_result)
messages
```

Combining all

```python
llm_with_tools.invoke(messages)
```
